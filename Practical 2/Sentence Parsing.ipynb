```python
sys_calls = []
for j in sentences:
    feature_vector = []
    for i in j:
        if "<" in i:
            call = i[4:]
            call = call.replace("<", "")
            call = call.replace("/>", "")
            try:
                feature_vector.append(hash_table[call])
            except:
                pass
    sys_calls.append(feature_vector)
```

```python
features = pd.DataFrame(sys_calls)
```

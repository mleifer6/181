```python
#PARAMS
direc="train"
num_trains = int(4*len(os.listdir(direc))/5.0)
num_valids = int(len(os.listdir(direc))/5.0)
full = int(len(os.listdir(direc)))
full_test = int(len(os.listdir("test")))
```

```python
import xml.etree.ElementTree as ET
import classification_starter as start
import os
from collections import Counter
try:
    import xml.etree.cElementTree as ET
except ImportError:
    import xml.etree.ElementTree as ET
import numpy as np
from scipy import sparse

import util
```

```python
sentences = []
for i in range(full):
    tree = ET.parse(os.path.join(direc,os.listdir(direc)[i]))
    tree_string = ET.tostring(tree.getroot(), method='xml')
    tree_string = tree_string.replace('=', " ")
    sentences.append(tree_string.split(' '))
```

```python
sentences_test = []
for i in range(full_test):
    tree = ET.parse(os.path.join("test",os.listdir("test")[i]))
    tree_string = ET.tostring(tree.getroot(), method='xml')
    tree_string = tree_string.replace('=', " ")
    sentences_test.append(tree_string.split(' '))
```

```python
sentences_valid = []
for i in range(num_trains,num_trains+num_valids):
    tree = ET.parse(os.path.join(direc,os.listdir(direc)[i]))
    tree_string = ET.tostring(tree.getroot(), method='xml')
    tree_string = tree_string.replace('=', " ")
    sentences_valid.append(tree_string.split(' '))
```
